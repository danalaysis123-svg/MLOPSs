```python
import pandas as pd
```

```python
employee = pd.read_csv(
    "employee.csv",
    sep=r"\s{2,}",
    engine="python"
)
```

```python
print(employee.columns)
```

```python
print(employee.head())
```

```python
expense = employee.groupby("Department")["Salary"].sum()

print(expense)
```

```python
import matplotlib.pyplot as plt

colors = {
    "R(CSE)": "red",
    "B(ECE)": "blue",
    "Y(EEE)": "yellow"
}

bar_colors = [colors[d] for d in expense.index]

plt.figure(figsize=(7,5))
plt.bar(expense.index, expense.values, color=bar_colors)

plt.title("Employee Salary Expense by Department")
plt.xlabel("Department")
plt.ylabel("Total Salary Expense")

plt.show()
```
