```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- MOCK DATA SETUP ---
# Creating dummy structures matching your problem inputs
employee_raw = {
    'Department': ['CSE', 'ECE', 'EEE', 'CSE', 'ECE'],
    'Employee Salary': [90000, 75000, 65000, 85000, 70000]
}

student_raw = {
    'Department': ['EEE', 'CSE', 'ECE', 'CSE', 'EEE'],
    'Total Fee': [120000, 150000, 130000, 160000, 110000]
}

# Convert to DataFrames
df_emp = pd.DataFrame(employee_raw)
df_stu = pd.DataFrame(student_raw)

# --- PIPELINE PROCESSING STEPS ---

# Step 1: Grab the data from employee for total expense
expense_df = df_emp.groupby('Department')['Employee Salary'].sum().reset_index()
expense_df.columns = ['Department', 'Total Expense']

# Step 2: Grab the data from student for total fee
fee_df = df_stu.groupby('Department')['Total Fee'].sum().reset_index()
fee_df.columns = ['Department', 'Total Fee']

# Step 3: Combine and sort the data according to department alphabetically
merged_df = pd.merge(expense_df, fee_df, on='Department')
merged_df = merged_df.sort_values(by='Department').reset_index(drop=True)

print("--- Cleaned, Sorted Department Data ---")
print(merged_df)

# --- STEP 4: PLOT IT ACCORDING TO DEPARTMENT ---
# Reorganizing data layout slightly to make plotting with Seaborn clean
melted_df = pd.melt(merged_df, id_vars=['Department'],
                    value_vars=['Total Expense', 'Total Fee'],
                    var_name='Financial Metric', value_name='Amount')

plt.figure(figsize=(8, 5))
# Custom colors mapping roughly to your structural prompt references
custom_colors = {'Total Expense': '#e74c3c', 'Total Fee': '#3498db'}

sns.barplot(data=melted_df, x='Department', y='Amount',
            hue='Financial Metric', palette=custom_colors)

plt.title("University Department Analysis: Expense vs Fee Revenue", fontsize=14, fontweight='bold')
plt.xlabel("Department", fontsize=12)
plt.ylabel("Amount (in ₹ / $)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Prepare the data from your Employee Management System output
# X = Experience (Independent variable) - Shaped as a 2D array for scikit-learn
X = np.array([5, 5, 2, 10, 11]).reshape(-1, 1)

# y = Salary (Dependent variable)
y = np.array([1500, 14000, 5000, 50000, 25000])

# 2. Initialize and fit the Linear Regression Model
model = LinearRegression()
model.fit(X, y)

# 3. Generate predictions for the line of best fit
X_predict = np.linspace(X.min() - 1, X.max() + 1, 100).reshape(-1, 1)
y_predict = model.predict(X_predict)

# 4. Plotting the results
plt.figure(figsize=(10, 6))

# Plot the actual employee data points
plt.scatter(X, y, color="blue", s=100, zorder=5, label="Actual Employees")

# Plot the regression line
plt.plot(
    X_predict,
    y_predict,
    color="red",
    linewidth=2,
    label=f"Regression Line (R² = {model.score(X, y):.2f})",
)

# 5. Graph Customization
plt.title("Employee Salary vs. Experience (Linear Regression)", fontsize=14)
plt.xlabel("Experience (Years)", fontsize=12)
plt.ylabel("Salary ($)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# 6. Display the graph window in PyCharm
plt.tight_layout()
plt.show()

# Print the mathematical formula to the console
print(f"Model Equation: Salary = {model.coef_[0]:.2f} * Experience + {model.intercept_:.2f}")
```
